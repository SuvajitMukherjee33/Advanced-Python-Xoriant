```python
def greet(name='', age=25):
    """
    This function is responsible to greet people
    """
    print("hello{}".format(name))
    print("WElcome to python")
    if isinstance(age,(int,float)):
        print("age",age)

greet(name="Prajakta")
```

```python
def greeting(name,age,phone="0000000",address="Pune"):
    print("name {}".format(name))
    print("age {}".format(age))
    print("phone {}".format(phone))
    print("address {}".format(address))

greeting("Praj",21)
```

```python
def greeting(*args,phone="0000000",address="Pune"):
    print("complete args{}"format(args))
    print("Hello {}".format(args[0]))
    print("age {}".format(args[1]))
    print("phone "phone)
    print("address"address)

greeting("Praj",21,phone="7687267",address="Manchar")
```

```python
def get_data(name,*args,phone="000",address="hdj"):
    print("name:",name)
    print("args:",args)
    print("phone:",phone)
    print("Address:",address)
get_data("Praj","IT","Pune",phone="87788",address="Pune")
```
