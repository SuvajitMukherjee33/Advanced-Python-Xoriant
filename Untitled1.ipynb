```python
def get_data(name,*args,phone="000",address="hdj"):
    print("name:",name)
    print("args:",args)
    print("phone:",phone)
    print("Address:",address)
get_data("Praj","IT","Pune",phone="87788",address="Pune")
```

```python
def greeting(*args,phone="0000000",address="Pune"):
    print("complete args{}".format(args))
    print("Hello {}".format(args[0]))
    print("age {}".format(args[1]))
    print("phone ",phone)
    print("address",address)

greeting("Praj",21,phone="7687267",address="Manchar")
```

```python
def get_data(name,*args,**kwargs): #kwargs=key words parameter like phone="988999"
    print("name:",name)
    print("args:",args)
    print("kwargs:",kwargs)
get_data("Praj",868493,"IT","Pune", phone="988999",location="Pune")
```

```python
def get_data(name,*args,discount=100,**kwargs): #kwargs=key words parameter like phone="988999"
    print("name:",name)
    print("args:",args)
    print("phone:",kwargs)
    print("discount:",discount)
get_data("Praj",8799,"IT","Pune", phone="988999",location="Pune")
```
