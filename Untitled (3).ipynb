```python
def fib(n):
    numbers=[]
    first,second=0,1
    while first < n:
        numbers.append(first)
        first,second=second,first+second
    print(numbers)
```

```python

##custom generator function like range
def frange(start, stop, increment):
    x=start
    while x< stop:
        yield x
        x+=increment

c=frange(0,4,0.5)
print(type(c))

for n in frange(0,4,0.5):
    print (n)

```

```python
#zip iterator
x=[1,2,3,45,6]
y=[101,102,103,104,109,38,47]
d={}
for i, j in zip(x,y):
    d[i]=j
print(d)
```

```python
# flattening the nested sequence
from collections import Iterable
def flatten(items, ignore_types={str,bytes}):
    for x in items:
        if isinstance(x,Iterable)and not isinstance(x,ignore_types):
            yield from flatten(x)
        else:
            yield x

items=[1,2,3,[4,5,[99,48],6],77]

for x in flatten(items):
    print(x)
            
```
