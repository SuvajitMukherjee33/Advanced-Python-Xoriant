```python
#general purpose decorators

def a_decorator_passing_arbitory_arguments(function_to_decorate):
    def a_wrapper_accepting_arbitory_arguments(*args,**kwargs):
        print("positional arguments are",args)
        print("Keyword arguments are",kwargs)
        function_to_decorate(*args)
    return a_wrapper_accepting_arbitory_arguments

@a_decorator_passing_arbitory_arguments
def function_with_no_arguments():
    print("No arguments here...")
    
#calling
function_with_no_arguments()
```

```python
#passing positional arguments to decorator
def accept_the_arguments(function_to_decorator):
    def inner_function(*args,**kwargs):
        print("positional arguments are",args)
        print("Keyword arguments are",kwargs)
        function_to_decorator(*args)
    return inner_function

@accept_the_arguments
def function_with_args(a,b,c):
    print(a,b,c)

#calling
function_with_args(10,20,30)
```
